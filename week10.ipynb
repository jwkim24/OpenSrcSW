```python
#실습문제 4번
while 1 :
    print("메뉴를 선택해주세요 :")
    menu = input("1. 암호화\t2. 복호화\t3. 종료: ")
    
    if menu=="1":
        string = input("암호화할 문자열을 입력해주세요 : ")
        string.strip()
        temp = string.split()
        for i in temp[::-1]:
            for k in i :
                code = ord(k)+1
                print(chr(code),end='')
            print(end=' ')
            
    if menu=="2":
        string = input("복호화할 문자열을 입력해주세요 : ")
        string.strip()
        temp = string.split()
        for i in temp[::-1]:
            for k in i :
                code = ord(k)-1
                print(chr(code),end='')
            print(end=' ')
    if menu=="3":
        break
```

```python
#실습문제 5번
print("전화번호부를 만들어주세요")
book = dict()
for i in range(10):
    name = input("이름? ")
    num = input("전화번호? ")
    book[name]=num

while 1 :
    print("메뉴를 선택해주세요 :")
    menu = input("1. 이름으로 검색\t2. 전화번호로 검색\t3. 종료: ")
    
    if menu=="1":
        key = input("이름? ")
        key.strip()
        if key in book:
            print(key,"의 전화번호는 ",book.get(key))
        else :
            print(key,"라는 사용자가 존재하지 않습니다.")
    if menu=="2":
        key = input("전화번호? ")
        if key in book.values():
            for i in book.keys():
                if book.get(i) == key:
                    print(key,"는 ",i,"의 전화번호입니다.")       
        else :
            print(key,"는 존재하지 않는 번호입니다.")
    if menu=="3":
        break
```
