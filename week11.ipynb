```python
#실습 7번 문제

userinput = input()
userinput = userinput.split()

if userinput[0] == "wc":
    fileName = userinput[1]

    f = open(fileName,'rt', encoding='UTF8')
    lines = f.readlines()  #각 줄을 요소로 갖는 리스트로 얻음
    f.close()

    print("# of lines : ",len(lines))  #리스트의 길이가 줄의 수

    wordNum = 0

    for line in lines:
        words = line.split()  #공백을 구분으로 개수 세줌
        wordNum += len(words)
    
    print("# of words : ",wordNum)
```

```python
#실습 8번 문제

#Calc클래스 정의
# -*-coding: cp949 -*-

class Calc:
    def sum(self, a, b) :
        result = a + b
        print("{0:d} + {1:d} = {2:d} 입니다.".format(a,b,result))
    def sub(self, a, b) :
        result = a - b
        print("{0:d} - {1:d} = {2:d} 입니다.".format(a,b,result))
    def multi(self, a, b) :
        result = a * b
        print("{0:d} * {1:d} = {2:d} 입니다.".format(a,b,result))
    def divi(self, a, b) :
        result = a / b
        print("{0:d} / {1:d} = {2:.0f} 입니다.".format(a,b,result))

#메인 테스트        
calc=Calc()
calc.sum(1,2)
calc.sub(5,1)
calc.multi(2,3)
calc.divi(6,2)
```

```python
#실습 9번 문제
#스택클래스 구현
class Stack:
    
    def __init__ (self):
        self.stack = []
    def size (self):
        return len(self.stack)
    def push(self, data):
        self.stack.append(data)
    def pop(self):
        if len(self.stack)<=0:
            return("NullPoint")
        else:
            temp = self.stack.pop()
            return temp

#메인
informula = input("계산식 입력 : ")
formula = informula.split()

num = Stack()
op = Stack()
k=0  #다음 수식이 *,/인지 확인해주는 변수
for i in formula :
        
        if i=="+" or i=="-" :
            op.push(i)
        elif i=="*" or i=="/":
            op.push(i)
            op.pop()  #뒤에서 처리해줄 것이기 때문에 스택에서 없애줌
        else :
            temp = int (i)
            num.push(temp)
        
        k+=1  #다음 수식이 *,/인지 확인해주기 위해 인덱스 변경
        
        if k<len(formula):   
            #곱하기와 나누기인 경우 저장되어 있던 num과 다음 숫자를 먼저 계산해주고 결과 push
            if formula[k]=="*" :
                a = num.pop()
                b = int (formula[k+1])
                del formula[k+1]  #계산했으니 리스트에서 삭제해줌 
                result = a * b
                num.push(result)
            elif formula[k]=="/" :
                a = num.pop()
                b = int (formula[k+1])
                del formula[k+1]
                result = a / b
                num.push(result)
                
        if num.size()>=2 :
            #num스택에 2개이상 있는경우 계산해줌
            temp = op.pop()
            if temp == "+" :
                c = num.pop()
                d = num.pop()
                result = (c + d)
                num.push(result)
            elif temp == "-" :
                c = num.pop()
                d = num.pop()
                result = (d - c)    
                num.push(result)
                
print("계산식 출력 : ",informula," = ",num.pop())
```
