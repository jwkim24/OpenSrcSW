```python
#실습 문제 1번

import matplotlib.pyplot as plt
import csv

f=open('seoul.csv')
data = csv.reader(f)
next(data)
high = []  #최고기온
low = []   #최저기온
for row in data :
    if row[-1] != '' and row[-2] != '':
        #생일인 10/26의 데이터만 가져오기
        if row[0].split('-')[1] =='10' and row[0].split('-')[2] == '26':  
            high.append(float(row[-1])) #최고기온 저장
            low.append(float(row[-2])) #최저기온 저장

f.close()
#그래프 색과 라벨 지정
plt.plot(high, color='pink',label='high') 
plt.plot(low, color='skyblue',label='low')
plt.legend()
plt.show()
```

```python
#실습 문제 2번

import matplotlib.pyplot as plt
import csv

f=open('seoul.csv')
data = csv.reader(f)
next(data)
aug = [] #8월 최고기온
jan = [] #1월 최고기온
for row in data :
    if row[-1] != '':
        if row[0].split('-')[1] =='01':  #1월인지 확인
            jan.append(float(row[-1])) #최고기온 저장
        elif row[0].split('-')[1] =='08':  #8월인지 확인
            aug.append(float(row[-1])) #최고기온 저장

f.close()
#히스토그램으로 표현
plt.hist(aug, bins=100, color='r', label = 'Aug')
plt.hist(jan, bins=100, color='b', label = 'Jan')
plt.legend()
plt.show()
```
